```python
#dependency 
import pandas as pd

#load file - didn't bother with putting it into a resources folder
load = "purchase_data.csv"

#read and store the load file to pd 
purchase_data = pd.read_csv(load)
purchase_data.head(2)
```

## Player Count

- Display the total number of players

```python
# value_count column "SN" and use the length function for Total PLayers
total_players = purchase_data["SN"].value_counts()
total_players = len(total_players)
total_players
```

```python
# create data frame of Total PLayers using a list of dictionaries
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)
- Run basic calculations to obtain number of unique items, average price, etc.
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame


```python
# calculate unique items, average price, number of purchases and total revenue
unique_items = purchase_data["Item Name"].value_counts()
unique_items = len(unique_items)
# print(unique_items) as a test = 179
average_price = purchase_data["Price"].mean()
# print(average_price) as a test = 3.05
purchase_total = purchase_data["Purchase ID"].value_counts()
purchase_total = len(purchase_total)
# print(purchase_total) as a test = 780
total_revenue = purchase_data["Price"].sum()
# print(total_revenue) as a test = 2379.77

# create a data frame using a list of dictionaries/values 
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                       
                                       "Average Price": [average_price], 
                                       
                                       "Number of Purchases": [purchase_total],
                                       
                                       "Total Revenue": [total_revenue]})

# give the displayed data cleaner formatting / have to use map to format columns $ & decimal because format will not work on a float object
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:,.2f}".format)


purchasing_analysis_df
```

## Age Demographics
 
- Establish bins for ages
- Categorize the existing players using the age bins. Hint: use pd.cut()
- Calculate the numbers and percentages by age group
- Create a summary data frame to hold the results
- Optional: round the percentage column to two decimal points
- Display Age Demographics Table

```python
# determine max age
max_age = purchase_data["Age"].max()
#print(max_age) = 45

# add bins and label them
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
label_age_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Place the data series into a new column inside of the DataFrame
purchase_data["View Age Group"] = pd.cut(purchase_data["Age"], bins = age_bins, labels = label_age_bins) 
#purchase_data

# Create a GroupBy object based upon "View Age Group"
age_group_df = purchase_data.groupby("View Age Group")

# count unique ages from SN
age_count_SN = age_group_df["SN"].nunique()

# age percentages for each age 
age_pct = age_count_SN / total_players * 100

# data frame summary
age_demographics_df = pd.DataFrame({"Total Count": age_count_SN, "Percentage of Players": age_pct})

# give the displayed data cleaner formatting
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

# hide the "View Age Group" column title
age_demographics_df.index.name = None 
age_demographics_df
```

- Bin the purchase_data data frame by age

## Top Spenders
 
- Run basic calculations to obtain the results in the table below
- Create a summary data frame to hold the results
- Sort the total purchase value column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
# groupby DF and calculate count/avg/total
TS_group_df = purchase_data.groupby("SN")
TS_purchase_count = TS_group_df["Purchase ID"].count()
TS_avg_purchase_price = TS_group_df["Price"].mean()
TS_total_purchase_value = TS_group_df["Price"].sum()

# summary dataframe
top_spenders_df = pd.DataFrame({"Purchase Count": TS_purchase_count,
                                "Average Purchase Price": TS_avg_purchase_price,
                                "Total Purchase Value": TS_total_purchase_value})

# sort the summary df to decend from high to low
top_spenders_sort= top_spenders_df.sort_values(["Total Purchase Value"], ascending = False)
#top_spenders_sort

# give the displayed data cleaner formatting
top_spenders_sort["Average Purchase Price"] = top_spenders_sort["Average Purchase Price"].map("${:.2f}".format) 
top_spenders_sort["Total Purchase Value"] = top_spenders_sort["Total Purchase Value"].map("${:.2f}".format) 

top_spenders_sort.head()
```

## Most Popular Items
- Retrieve the Item ID, Item Name, and Item Price columns
- Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
- Create a summary data frame to hold the results
- Sort the purchase count column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
# groupby df and calculate count/total/item price
most_popular_group_df = purchase_data.groupby(["Item ID", "Item Name"])
MP_purchase_count = most_popular_group_df["Purchase ID"].count()
MP_total_purchase_value = most_popular_group_df["Price"].sum()
MP_item_price = MP_total_purchase_value / MP_purchase_count 

# dataframe summary
most_popular_df = pd.DataFrame ({"Purchase Count": MP_purchase_count,
                                 "Item Price": MP_item_price,
                                 "Total Purchase Value": MP_total_purchase_value})

# Sort the summary df to descend from high to low
most_popular_df_sort = most_popular_df.sort_values(["Purchase Count"], ascending = False)

# give the displayed data cleaner formatting
most_popular_df_sort["Item Price"] = most_popular_df_sort["Item Price"].map("${:.2f}".format)
most_popular_df_sort["Total Purchase Value"] = most_popular_df_sort["Total Purchase Value"].map("${:.2f}".format)

most_popular_df_sort.head(10)
```
